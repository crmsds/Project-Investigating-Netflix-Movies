```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
file_path = '/content/drive/MyDrive/GitHub/Project-Investigating-Netflix-Movies/netflix_data.csv'
netflix_df = pd.read_csv(file_path)

# Filter the data to remove TV shows and store as netflix_subset
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']

# Investigate the Netflix movie data
selected_columns = ["title", "country", "genre", "release_year", "duration"]

# Create a new DataFrame with only the selected columns
netflix_movies = netflix_subset[selected_columns]

# Filter netflix_movies to find the movies that are shorter than 60 minutes
short_movies = netflix_movies[netflix_movies['duration'] < 60]

# Initialize an empty list to store colors
colors = []

# Iterate through rows of netflix_movies and assign colors based on genre
for index, row in netflix_movies.iterrows():
    genre = row['genre']
    if genre == 'Children':
        colors.append('blue')
    elif genre == 'Documentaries':
        colors.append('green')
    elif genre == 'Stand-Up':
        colors.append('red')
    else:
        colors.append('gray')  # Assign 'Other' genre movies to gray

# Create a scatter plot
fig, ax = plt.subplots()

# Scatter plot with colors
scatter = ax.scatter(netflix_movies['release_year'], netflix_movies['duration'],
                     c=colors, alpha=0.5)

# Add labels and title
ax.set_xlabel('Release year')
ax.set_ylabel('Duration (min)')
ax.set_title('Movie Duration by Year of Release')

# Show the plot
plt.show()

answer = "maybe"
```
